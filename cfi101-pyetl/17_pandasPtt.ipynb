```python
import pandas as pd
import glob

glob.glob('./res_gossiping/*.txt')
```

```python
data = list()

for filePath in glob.glob('./res_gossiping/*.txt'):
    with open(filePath, 'r', encoding='utf-8') as f:
        data.append(f.read().split('---split---')[1].split('\n')[1:-1])

data
```

```python
columns = [r.split(': ')[0] for r in data[0]]
columns
```

```python
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
df2 = df.copy()
```

```python
def columnFilter(s):
    result = s.split(': ')[1]
    return result
```
