```python
import jieba
import os

os.listdir('./res_gossiping')
jieba.load_userdict('./mydict.txt')
stopwordList = list()
with open('./stopword.txt', 'r', encoding='utf-8') as f:
    stopwordList = [w.replace('\n', '') for w in f.readlines()]
stopwordList
```

```python
sb = ''
for fileName in os.listdir('./res_gossiping'):
    filePath = './res_gossiping/' + fileName
    with open(filePath, 'r', encoding='utf-8') as f:
        sb += f.read().split('---split---')[0]
        
sCut = jieba.cut(sb)
```

```python
wordCutList = [w for w in sCut]
wordCutList
```

```python
wordCountDict = dict()
for word in wordCutList:
    if word.__len__() > 1 word not in stopwordList:
        if word in wordCountDict:
            wordCountDict[word] += 1
        else:
            wordCountDict[word] = 1
```

```python
wordCountList = [(k, v) for k, v in wordCountDict.items()]
wordCountList
```
